```python
import pandas as pd
import seaborn as sns

df3 = pd.read_json("PI3_com-sem_aquecimento/READ_WRITE/5m_PEGA_ESSE_AQUI__READ-WRITE__256T__500W-1500R__JSON__.json")
```

```python
from matplotlib import pyplot as plt

plt.figure()
var1 = df3.reset_index().plot.line(x="currentTime", y=["writeLatP95", "writeLatP99", "writeLatAvg"])
var1.set_xticks([0, 300, 600, 900, 1200])
var1.set_xticklabels([5,10, 15, 20, 25])
var1.set_yticklabels([0, 500, 1000, 1500, 2000, 2500, 3000])
var1.set_xlabel("Tempo de Experimento (Minutos)")
var1.set_ylabel(u"Latncia de Escrita (ms)")
var1.legend(["Lat. P95", "Lat. P99", "Lat. Media"])
plt.savefig("PI3__WRITE__READ-WRITE__LATENCIAS__500W-1500R", dpi=500 )

var2 = df3.reset_index().plot.line(x="currentTime", y=["readLatP95", "readLatP99", "readLatAvg"])
var2.set_xticks([0, 300, 600, 900, 1200])
var2.set_xticklabels([5,10, 15, 20, 25])
var2.set_yticklabels([0, 500, 1000, 1500, 2000, 2500])
var2.set_xlabel("Tempo de Experimento (Minutos)")
var2.set_ylabel(u"Latncia de Leitura (ms)")
var2.legend(["Lat. P95", "Lat. P99", "Lat. Media"])
plt.savefig("PI3__READ__READ-WRITE__LATENCIAS__500W-1500R", dpi=500 )

var2 = df3.reset_index().plot.line(x="currentTime", y=["writeRPS", "readRPS"])
var2.set_xticks([0, 300, 600, 900, 1200])
var2.set_xticklabels([5,10, 15, 20, 25])
# var2.set_yticklabels([0, 500, 1000, 1500, 2000, 2500])
var2.set_yticks([0, 150, 300, 450, 600])
var2.set_xlabel("Tempo de Experimento (Minutos)")
var2.set_ylabel(u"Requisio por segundo")
var2.legend(["Req. Escrita", "Req. Leitura"])
plt.savefig("PI3__RPS__READ-WRITE__LATENCIAS__500W-1500R", dpi=500 )
```
